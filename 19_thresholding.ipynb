```python
ret_value1,thres_img1 = cv2.threshold(gray,75,255,cv2.THRESH_BINARY_INV)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret_value2,thres_img2 = cv2.threshold(gray,75,255,cv2.THRESH_MASK)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret_valu3,thres_img3 = cv2.threshold(gray,75,255,cv2.THRESH_OTSU)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to zero  : any value greatern than 71 ==1 , less than 75 ==0
ret_value4,thres_img4 = cv2.threshold(gray,75,255,cv2.THRESH_TOZERO)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to inv zero  : any value greatern than 71 ==1 , less than 75 ==0
ret_value5,thres_img5= cv2.threshold(gray,75,255,cv2.THRESH_TOZERO_INV)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# threshold triangulation
ret_value5,thres_img5 = cv2.threshold(gray,75,255,cv2.THRESH_TRIANGLE)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# threshold truncattion
ret_value6,thres_img6 = cv2.threshold(gray,75,255,cv2.THRESH_TRUNC)
cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img6)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret_value6,thres_img6 = cv2.threshold(gray,75,255,cv2.THRESH_TRUNC)
ret_value5,thres_img5 = cv2.threshold(gray,75,255,cv2.THRESH_TRIANGLE)
ret_value5,thres_img5= cv2.threshold(gray,75,255,cv2.THRESH_TOZERO_INV)
ret_value4,thres_img4 = cv2.threshold(gray,75,255,cv2.THRESH_TOZERO)
ret_valu3,thres_img3 = cv2.threshold(gray,75,255,cv2.THRESH_OTSU)
ret_value2,thres_img2 = cv2.threshold(gray,75,255,cv2.THRESH_MASK)
ret_value1,thres_img1 = cv2.threshold(gray,75,255,cv2.THRESH_BINARY_INV)

cv2.imshow('gray',gray)
cv2.imshow('threshold_img',thres_img1)
cv2.imshow('threshold_img',thres_img2)
cv2.imshow('threshold_img',thres_img3)
cv2.imshow('threshold_img',thres_img4)
cv2.imshow('threshold_img',thres_img5)
cv2.imshow('threshold_img',thres_img6)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
adapt_avg= cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,
                                 cv2.THRESH_BINARY,
                                3,1)
```

```python
cv2.imshow('Adapt averate',adapt_avg)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
adapt_avg2= cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,
                                 cv2.THRESH_BINARY,
                                11,5)
cv2.imshow('Adapt averate',adapt_avg2)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
adapt_gaus = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                     cv2.THRESH_BINARY,
                                13,3)
cv2.imshow('Adapt averate',adapt_avg2)
cv2.imshow('Adapt gausian',adapt_gaus)

cv2.waitKey(0)
cv2.destroyAllWindows()

```
