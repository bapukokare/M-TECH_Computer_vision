Real time detection with dnn with video

```python
import numpy as np
import cv2
```

```python
#image = img.copy()
#create face detection function
def face_detection_dnn(img):
    #step 1: blob form image
    blob = cv2.dnn.blobFromImage(img,1,(300,300),(104,177,123),swapRB=True)
    #step2 : set blob as input
    face_detection_model.setInput(blob)
    #step3 : get the output
    detections = face_detection_model.forward()
    #step 4: drawing bounding box on top o face detected
    image = img.copy()
    
    h,w = image.shape[:2]
    for i in range(0,detections.shape[2]):
        confidence = detections[0,0,i,2]
        #print(confidence)
        if confidence > 0.5:
            print(confidence)
            # diagnonal point3:7 
            box = detections[0,0,i,3:7]*np.array([w,h,w,h])
            box = box.astype('int')
            pt1 = (box[0],box[1])
            pt2 = (box[2],box[3])
            #print(box)
            # draw rectangle
            cv2.rectangle(image,pt1,pt2,(0,255,0),1)
            text = 'score:{:.0f}'.format(confidence*100)
            print(text)
            cv2.putText(image,text,pt1,cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
    return image

```

```python
cap = cv2.VideoCapture(0)
face_detection_model = cv2.dnn.readNetFromCaffe('./model/deploy.prototxt.txt',
                                                './model/res10_300x300_ssd_iter_140000_fp16.caffemodel')

while True:
    ret,frame =cap.read()
    if ret == False:q
        break
    image_detection = face_detection_dnn(frame)
    cv2.imshow('Real Time Face detection with DNN', image_detection)
    if cv2.waitKey(1)== ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
